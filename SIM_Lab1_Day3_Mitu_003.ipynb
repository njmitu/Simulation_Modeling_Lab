```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

Assignment 1: Basic Matrix Operations

1. Create two 3Ã—3 matrices with random integers in Python.

2. Perform addition, subtraction, and multiplication operations on these matrices.

3. Compute the determinant, inverse, and rank of one of the matrices.

```python
A = np.random.randint(1, 11, size=(3, 3))
B = np.random.randint(1, 11, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)


matrix_sum = A + B          # Matrix addition
matrix_diff = A - B         # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication (dot product)

det_A = np.linalg.det(A)
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is Singular, cannot compute"
rank_A = np.linalg.matrix_rank(A)

print("Addition of matrix: \n",matrix_sum)
print("Subtraction of matrix: \n",matrix_diff)
print("Multiplication of matrix: \n",matrix_prod)

print("\ndet_A:\n", det_A)
print("\ninv_A:\n", inv_A)
print("\nrank_A:\n", rank_A)
```

1. Generate and Plot Two Random Vectors:


a. Create two vectors, each containing 15 random floats of your choice.

b. Plot these vectors on the same graph using matplotlib to compare how
they look.


c. Label the axes appropriately and give the plot a title.

```python
v1 = np.random.uniform(1, 11, 15)
v2 = np.random.uniform(0, 11, 15)

print("\nV1:\n", v1)
print("\nV2:\n", v2)

#Plot these vectors on the same graph using matplotlib
plt.figure()
plt.plot(v1, marker='o', linestyle='-', label="V1")
plt.plot(v2, marker='s', linestyle='--', label="V2")

# Label the axes
plt.xlabel("Index")
plt.ylabel("Value")

#Add a Title
plt.title('Comparison of Two Random Vectors')

# Add legend
plt.legend()
plt.grid(True)

plt.show()
```

2. Generate and Visualize a Random Matrix:

a. Create a 4x4 matrix with random values of your choice.

b. Display the matrix using a heatmap with appropriate color mapping (e.g.,
seaborn or matplotlib).

c. Add labels to the heatmap to indicate row and column numbers.

```python
# Generate a 4x4 matrix with random values
matrix = np.random.rand(4, 4)
print("Matrix:\n", matrix)

# Plot the heatmap
sns.heatmap(matrix, annot=True, cmap='YlGnBu', linewidths=1, linecolor ='black', xticklabels=['col 1', 'col 2', 'col 3', 'col 4'],
            yticklabels=['row 1', 'row 2', 'row 3', 'row 4'])

plt.title('Heatmap of 4x4 Matrix')
plt.show()
```

3. Matrix Operations and Visualization:

a. Create two 4x4 matrices with random values of your choice.

b. Perform the following operations:

i. Matrix Addition

ii. Matrix Subtraction

iii. Matrix Multiplication (ensure matrices are compatible for
multiplication)


c. For each of these operations, visualize the result using bar plots. Each bar
should represent one element of the matrix.


```python
# Create two random 4x4 matrices with integer values between 1 and 11
A = np.random.randint(1, 11, size=(4, 4))  # Random integers between 1 and 11
B = np.random.randint(1, 11, size=(4, 4))  # Random integers between 1 and 11

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix operations
matrix_sum = A + B          # i.Matrix addition
matrix_diff = A - B         # ii.Matrix subtraction
matrix_prod = np.dot(A, B)  # iii.Matrix multiplication (dot product)

# Bar plot to visualize results
fig, axes = plt.subplots(1, 3, figsize=(15, 5)) #figure object, subplot array


# Sum
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='teal')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='pink')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='plum')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout() #plt.tight_layout() is a Matplotlib function that automatically adjusts the layout of subplots and figures to prevent overlapping and ensure a visually appealing presentation. It modifies the spacing between subplots, axes, labels, and titles to improve readability.
plt.show()
```
